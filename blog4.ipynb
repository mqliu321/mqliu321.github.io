# Classification

## Introduction

**Classification** in machine learning is a type of **supervised learning** approach where the output variable is a category, such as "spam" or "not spam" or a diagnosis.

The main goal is to accurately predict the output class for each data point in the dataset. It helps in decision-making processes based on patterns learned from the data.

The classification mainly have two phases: 
- Training: A model is trained on a dataset containing features and corresponding labels.
- Testing: The trained model is tested on a new set of data to evaluate its performance.

The commonly used evaluation metrics include **accuracy, preision, recall, and F1 score**.

## Types of Classification

- **Binary Classification**: Classify an instance into two classes, such as determining if an email is spam or not.
- **Multiclass Classification**: Classify an instance into more than two classes.
- **Multilabel Classification**: Multiple labels are predicted for each instance.

## Common Classification Algotithms

- **Support Vector Machines (SVM)**: Finds a hyperplane in N-dimensional space that distinctly classifies the data points.
- **Decision Trees**: Tree-like model of decisions and their possible consequences.
- **Naive Bayes**: Based on Bayesâ€™ Theorem with an assumption of independence among predictors.
- **K-Nearest Neighbors (KNN)**
- **Neural Networks**

## Demonstration: SVM

IN this blog, we use SVM as an instance to show more details on how classification algorithm works.

### Introduction on SVM

**Support Vector Machines (SVM)** are a set of supervised learning methods used for classification, regression, and outlier detection. They are particularly well-known for their effectiveness in high-dimensional spaces and their versatility in handling both linear and non-linear data.

The following are the key concepts of SVM:
- **Hyperplane**: In SVM, the data are divided by a line or a plane (in higher dimensions). This line or plane is known as a hyperplane. The goal of the SVM algorithm is to find the optimal hyperplane which best separates the data into classes.
- **Margin**: The margin is the gap between the two lines on the closest class points. This is calculated as the perpendicular distance from the line to the support vectors or closest points. The optimal hyperplane is the one with the largest margin.
- **Support Vectors**: Support vectors are the data points that are closest to the hyperplane. These points are critical in defining the hyperplane and thus the SVM classifier.
- **Kernel Trick**: For non-linearly separable data, SVM uses a kernel trick to transform the input space to a higher dimensional space where it is easier to separate the data linearly. Common kernels include linear, polynomial, radial basis function (RBF), and sigmoid.


### Implementation and Visualization of SVM

In the following, we use the Iris dataset as an example to demonstrate the SVM algorithm. We plot the decision boundary of the trained SVM classifier.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix

iris = datasets.load_iris()
X = iris.data[:, :2]
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=6)

model = SVC(kernel='linear')
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))

plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap='winter')

x_min, x_max = X_train[:, 0].min() - 1, X_train[:, 0].max() + 1
y_min, y_max = X_train[:, 1].min() - 1, X_train[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))

Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, alpha=0.8, cmap='winter')


plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('SVM Classifier')
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.show()

```
