# Clustering

## Definition

**Clustering** is the task of grouping a set of objects such that the items in the same cluster are more similar (usually measured by a similarity metric) to each other tan to those in other clusters. 

Clustering is a fundamental topic in data analysis and machine learning, primarily used for pattern recognition and discovering meaningful structures within data. 

It's a type of **unsupervised learning**, which means the algorithm tries to find patterns without any pre-labeled outcomes to guide the process.

## Types of Clustering

We give a brief introduction to 4 typical types of clustering:

- **Connectivity-based clustering**: Connectivity-based clustering, aka hierarchical clustering, is based on the idea of objects being more related to nearby objects than to objects farther away. It provides an extensive hierarchy of clusters that merge with each other at certain distances.
- **Centroid-based clustering**: Each cluster is represented by a central vector (i.e., centroid), which is not necessarily a datapoint in the dataset. One representative algorithm of this type of clustering is K-means clustering, which has wide application in machine learning.
- **Distribution-based clustering**: The approach closely resembles the way artificial data sets are generated: by sampling random objects from a distribution.
- **Density-based clustering**: Clusters are defined as areas of higher density than the remainder of the dataset. Objects in sparse areas – that are required to separate clusters – are usually considered to be noise and outliners.

## K-Means Clustering

K-means is a type of partitional clustering algorithm. It's simple, efficient, and one of the most popular clustering methods.

### Overview of K-Means

- **Initialization**: Randomly pick K points as the initial centroids.
- **Assignment Step**: Assign each data point to the nearest centroid, creating K clusters.
- **Update Step**: Recalculate the centroids as the mean of all points in the cluster.
- **Iteration**: Repeat the assignment and update steps until convergence (i.e., when centroids do not significantly change between iterations).

### Implementation and Visulization of K-Means

Our demonstration includes 3 parts: 
- (1) Generate a toy dataset;
- (2) Implement K-Means algorithm;
- (3) Visualize results.



```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans

# Load Iris dataset
iris = datasets.load_iris()
X = iris.data[:, :2] 

# Implement K-Means Clustering
kmeans = KMeans(n_clusters=3)  # The Iris dataset has 3 species
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], s=50)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title("Initial Datapoints")

plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.5)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title("Datapoints after K-Means")

plt.show()

```
