# Anomaly/outlier detection

## Introduction


Anomaly or outlier detection in machine learning refers to the identification of rare items, events, or observations that raise suspicions by differing significantly from the majority of the data. This concept is crucial in various domains such as fraud detection, network security, fault detection, system health monitoring, event detection in sensor networks, and detecting ecosystem disturbances.


## Applications
- **Fintech Fraud Detection**: Identifying fraudulent activities in banking and financial systems.
- **Intrusion Detection**: Detecting intrusions and threats in network systems.
- **Health Monitoring**: Identifying unusual patient conditions in healthcare monitoring systems.
- **IT Infrastructure**
- **IoT Systems**

## Approaches
- **Statistical Approach**: prametric-based approach such as Z-score and parametric-free approach.
- **Density-based Approach**: examples include k-nearest neighbor (k-NN) and one-class support vector machines.
- **Neural Networks**: Autoencoders, LSTM, Hidden Markov models (HMMs), and deep neural networks.
- **Clustering-Based Approach**

## Implementation of k-Nearest Neighbor (k-NN) for Anomaly Detection

We create a toy dataset by randoming adding outlines. Then we implement a k-NN algorithm to detect the outliners. Finally, we visualize the results where the anomaly outliners are highlighted in red.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import NearestNeighbors

# Step 1: Create a Synthetic Dataset
np.random.seed(0)
normal_data = np.random.randn(100, 2)
outliers = np.random.uniform(low=-6, high=6, size=(10, 2))
data = np.vstack([normal_data, outliers])

# Step 2: Implement k-NN
k = 5
nbrs = NearestNeighbors(n_neighbors=k).fit(data)
distances, indices = nbrs.kneighbors(data)

# Step 3: Anomaly Detection
# Average distance to k neighbors for each point
avg_distance = np.mean(distances, axis=1)
threshold = np.percentile(avg_distance, 90)  # setting threshold as 90th percentile
anomaly_indices = np.where(avg_distance > threshold)[0]

# Step 4: Visualization
plt.figure(figsize=(10, 6))
plt.scatter(data[:, 0], data[:, 1], color='blue', label='Normal Data')
plt.scatter(data[anomaly_indices, 0], data[anomaly_indices, 1], color='red', label='Anomalies')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Anomaly Detection using k-NN')
plt.legend()
plt.grid(True)
plt.show()
```
